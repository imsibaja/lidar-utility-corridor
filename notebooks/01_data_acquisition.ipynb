# 01 — Data Acquisition and Point Cloud Inspection

## What this notebook does

This notebook covers the first two steps of the pipeline:
1. Acquiring a USGS 3DEP LiDAR tile (.laz) for the study area
2. Loading it into Python and inspecting its structure before any processing begins

No PDAL processing is performed here. This notebook is diagnostic only.

---

## Inputs expected

- A single `.laz` tile downloaded from **OpenTopography** (opentopography.org) or
  the **USGS National Map Downloader** (tnmaccess.usgs.gov).
- The tile should be QL1 (≤0.35 pts/m² ground, typically 8+ pts/m² total) or
  QL2 (≤0.71 pts/m² ground, typically 2+ pts/m² total). QL3 is insufficient
  for 1m CHM derivation.
- Place the downloaded file at: `data/raw/<tile_name>.laz`

## Outputs produced

- No files written. Diagnostic output only (printed summaries).

## src/ functions called

- `src.data_utils.load_laz(filepath)` — reads the .laz file
- `src.data_utils.inspect_point_cloud(las_data)` — prints return count,
  classification distribution, bounding box, and estimated point density
- `src.data_utils.validate_crs(las_data, expected_epsg)` — confirms the
  tile CRS matches the expected projected CRS before any downstream processing

---

## Key decisions to make before writing code

- **Study area selection:** The target tile is TBD. Use the OpenTopography
  map interface or the USGS National Map Availability Viewer to find a
  QL1/QL2 tile covering a California transmission corridor.
  Candidate regions: Ventura County, Santa Barbara County foothills,
  San Bernardino foothills.
- **Tile size:** Select a 1km² tile to keep file sizes manageable while
  learning the workflow.
- **Expected EPSG:** Confirm the horizontal CRS of the downloaded tile.
  USGS 3DEP is distributed in state plane or UTM. For Southern California,
  UTM Zone 11N (EPSG:26911) is a common target. Reproject if necessary
  before further processing.
- **Vertical datum:** Confirm NAVD88 before any elevation-dependent analysis.
  USGS 3DEP LiDAR uses NAVD88 by default, but verify for the specific tile.
